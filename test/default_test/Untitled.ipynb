```python
filename = "MultiSubworldSamperl_log"
import pandas as pd
import numpy as np
df = pd.read_csv(filename)
df = df.rename(columns={'Grid_y': 'Grid_Y'})
df = df.astype({'Grid_X' : 'int', 'Grid_Y':'int'})
```

```python
data = np.zeros(shape=(200,200), dtype ="int")
for idx, row in df.iterrows():
    data[int(row.Grid_X), int(row.Grid_Y)] = data[int(row.Grid_X), int(row.Grid_Y)] + 1
```

```python
import plotly.express as px

fig = px.imshow(data)
fig.show()
```

```python
df = pd.DataFrame(data.reshape((1,40000)))
df = df.transpose()
```
