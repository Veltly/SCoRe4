```python
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [16, 12]

df = pd.read_csv("depth_h1_depth.csv", header=6)
df_old = pd.read_csv("macros/shiftData", header=None)

x = [i * 125./1000. for i in range(0, len(df))] #rescale x

data = df['entries'].values
data = data / (data.sum() * len(df_old)/len(df)) #rescale data


x_old = df_old[0].values

y_old=df_old[1].values
y_old = y_old / y_old.sum()

fig, ax = plt.subplots()
ax.scatter(x_old,y_old,marker='x', label='template')
ax.scatter(x, data, marker='.',label='geant4 shift')
plt.yscale('log')
plt.legend()
plt.show()
```
