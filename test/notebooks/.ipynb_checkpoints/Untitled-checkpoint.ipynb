## Read logfile and extract data

```python
import re

PostStepPoint = "----DebugInfo----> PortalControl: PostStepPoint: x:"
SubworldPosition = " ----DebugInfo----> PeriodicPortal: Subworld:"

def read_all_lines_filter(contains, filename):
    with open(filename) as file:
        lines = file.readlines()
    coord = {'x':list(),'y':list(),'z':list(),'Volume':list()}
    subworld_coord = {'x':list(),'y':list()}
    for line in lines:

        if PostStepPoint in line:
            nums = re.findall(r'\d+\.?\d*',line)
            nums = [float(num) for num in nums]
            coord['x'].append(nums[0])
            coord['y'].append(nums[1])
            coord['z'].append(nums[2])
            volume = re.split("Volume is: ", line)
            if(len(volume) == 2):
                coord['Volume'] = volume[1]
        if SubworldPosition in line:
            nums = re.findall(r'\d+\.?\d*',line)
            nums = [int(num) for num in nums]
            subworld_coord['x'].append(nums[0])
            subworld_coord['y'].append(nums[1])
    return coord, subworld_coord
    
coord, subworld_coord = read_all_lines_filter("----DebugInfo---->" ,"log")
```

### Plot Data in pixelgrid

```python
import numpy as np
from PIL import Image

data = np.zeros( (1000,1000,3), dtype=np.uint8 )
for x in np.arange(1,1000,1):
    for y in np.arange(1,1000,1):
        data[x,y] = [240,240,240]

for x, y in zip(subworld_coord['x'], subworld_coord['y']):
    data[x,y] = [254,0,0]

img = Image.fromarray(data)
img.show()
```
